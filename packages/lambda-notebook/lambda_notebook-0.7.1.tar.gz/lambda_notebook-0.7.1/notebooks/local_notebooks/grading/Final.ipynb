I chose this type over type e since "there" is not a referential term in this sentence.

[a] The conative alternants seem to differ from their basic forms in their completion or successfulness. Adding "at" after a verb makes it seem less likely that the action was completed or sucessful. For example, "Alfonso banged at the door" or "Alfonso poked at the food" seems to indicate that Alfonso missed the door and the food, or at the very least, there is less certainty that these actions were sucessful. With "Alfonso cut at the bread," it seems like either he was unable to complete cutting the bread, or it took multiple attempts to cut (both of which indicate that the action was unsucessful, to varying degrees). As to "Alfonso chewed at the sushi," "at" seems to create the same effect as in the case with the verb "cut". It seems to indicate that he could not finish chewing the sushi (or that the chewing took longer than usual).


The difference in entailments between the basic form and the alternant is therefore, that the basic form entails completion/succesfulness of the action, while the alternant does not necessarily have this entailment. We can show that this is the case with the cancellation test. For example, the sentences: "Alfonso cut the bread" and "Alfonso licked the sushi". The entailments are that the bread has been cut and that Alfonso's tongue touched the sushi.

    (1a) #Alfonso cut the bread, but he could not cut it. 
    
    (2a) #Alfonso licked the sushi, but his tongue did not touch it.
    
     or: #Alfonso licked the sushi, but his tongue missed.
     
In both cases, neither sentence makes sense, so they are non cancellable and therefore entailments. However, when the basic forms are replaced with the conative alternants, they become cancellable: 

    (1b) Alfonso cut at the bread, but he could not cut it.
    
    (2b) Alfonso licked at the sushi, but his tongue missed.
    
Unlike 1a and 2a, both these scenarios make sense. They evoke images of someone performing an action (licking or handling a knife), but not making any or enough contact with an object. 

Therefore, the entailment for conative alternants is the process of the action, rather than its success or completion. For example:

    (1c) #Alfonso cut at the bread, but he did not make a cutting motion with a sharp object.
    
    (2c) #Alfonso licked at the sushi, but his tongue never left his mouth.
    
These sentences do not make sense, so the entailments for these sentences are that Alfonso performed some kind of cutting and licking motions.  
    






[b]

The argument structures of the basic form and conative form are different in that the conative will always be transitive, but the basic form can be intransitive, transitive, or ditransitive. For example, "chew" can exist in the intransitive and transitive forms:

    (1a) Chew
    
    (1b) Alfonso chews
    
    (1c) Alfonso chews the sushi

and "cut" can exist in the transitive and ditransitive forms:

    (2a) Alfonso cut the bread.
    
    (2b) Alfonso cut the bread into pieces.
    
However, by adding the word "at," both these verbs become only transitive:

    (3a) #Chew at
    
    (3b) #Alfonso chews at
    
    (3c) Alfonso chews at the sushi
    
    
    (4a) Alfonso cut at the bread.
    
    (4b) #Alfonso cut at the bread into pieces.
    
This is related to the aspectual class of the basic and conative forms. Some of the verbs, in their basic form, can be made to belong in any of the three classes (activity, accomplishment, achievement). In the sentences provided, most of the verbs seem to go from being either accomplishment or activity to being only activity as they go from basic form to conative form. For example:

    (3a) Alfonso cut the bread.
    
This phrase can be classified as an accomplishment if there is a specific, limited amount of bread being cut. In this case, the action has a defined endpoint (when the bread has finished being cut), and it happens over a time interval. It is also possible that there is no defined endpoint (for example, if Alfonso cuts bread for a living, and repeats this action for a specific time-interval, rather than until a specific amount of bread is cut), in which case it is an activity. 

We can see that this is the case since both "Alfonso cut the bread in an hour" and "Alfonso cut the bread for an hour" make sense.

However, in its conative form,

    (3b) Alfonso cut at the bread.
    
completion is not implied. There is therefore no defined endpoint, so it can only be an activity. We can show that this is true since "Alfonso cut at the bread in an hour" does not make sense, but "Alfonso cut at the bread for an hour" does. 



[c] The basic form and conative form have similar event structures. The basic form of a verb will follow a structure resembling: 

x cause y go to z

for example, "Alfonso cause knife go to (or through) bread", for the sentence "Alfonso cuts the bread".

The structure of the conative form of a verb, which is derived from the structure of the basic form, will follow the same structure, but replace its preposition with the preposition "at":

x cause y go at z

for example, "Alfonso cause knife go at bread", for "Alfonso cuts at the bread."

While the structures are similar, one difference is that the preposition in the basic form is incorporated into the verb, while the preposition in the conative form is not. This means that, in a syntax tree, the conative form will have an extra node. The VP node in a basic form tree will contain the verb ("cause y go to" or "cause knife go to") and a DP ("z" or "the bread"). The VP node in a conative form tree will contain the verb ("cause y go" or "cause knife go") and a prepositional phrase. The prepositional phrase node will then contain the preposition "at" and the same DP node as in the basic form. 

[d] extra credit

In order for a verb to work with the conative, it has to meet two main criteria. Firstly, as one can see from the examples provided, all the verbs are transitive action verbs. Non action verbs like "think" or "own", or intransitive verbs like "go" cannot work with "at". Secondly, the verb must describe a motion that results in contact. This criteria is made up of several parts. For one, the verb describes a motion. The verb "cut" describes a chopping or sawing motion with a (usually sharp) object, "lick" describes the motion of a tongue, "claw" describes the motion of one's fingers or fingernails. Verbs like "debate" or "sing" are transitive action verbs, but do not describe a motion, and will therefore not work with the conative (#Alfonso debates at John, #Alfonso sings at the song). Another part of this criteria is that the motion must result in contact. "Cut," "lick," and "claw" will all result in contact being made between the object/body part and another object. A verb like "launch" describes the motion of an object being thrown or propelled, but it does not necessarily involve contact (especially if the object is being launched into space), so it will not work with the conative (#Alfonso launched at a rocket). Finally, the criteria requires that the verb describes the motion that precedes the contact, not the motion of the contact. The verbs "cut," "lick," and "claw" all describe movement before contact is made. However, verbs like "touch," "rip," "smash," or "break," despite involving motion, are only concerned with the end result, not the motion leading up to it, and will therefore not work with the conative (#Alfonso touched at the wall, #Alfonso ripped at the paper, #Alfonso smashed at the dish, #Alfonso broke at the vase). Some verbs that follow these guidelines include: chop, bite, pull, and kick . These verbs are expected to work with the conative, which they do:

    (1) Alfonso chopped at the wood.
    (2) Alfonso bites at the sandwich.
    (3) Alfonso pulled at the door. 
    (4) Alfonso kicked at the door.
