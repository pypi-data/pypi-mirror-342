```python
Tree = lamb.utils.get_tree_class()
```

```python
lang.set_system(lang.hk3_system)
%lamb ||gray|| = L x_e : Gray_<e,t>(x)
%lamb ||cat|| = L x_e : Cat_<e,t>(x)
%lamb ||every|| = L f_<e,t> : L g_<e,t> : Forall x : f(x) >> g(x)
t = Tree.fromstring("(DP (D every) (NP (A gray) (N cat)))")
t2 = lang.CompositionTree.tree_factory(t)
r = lang.hk3_system.expand_all(t2)
```

```python
display(t, r.denotations[0].tree(style={'style': lamb.display.DerivStyle.PROOF}))
```
