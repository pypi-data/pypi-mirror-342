Code for tree diagrams illustrating relative clauses and movement:

```python
import svgling, svgling.figure
rctree = ("DP", ("D", "a"), ("NP", ("N", "doctor"),
                             ("CP", ("DP", "who(i)"),
                                    ("TP", ("DP", "t(i)"),
                                             ("VP", ("V", "likes"), ("DP", "cats"))))))
out = svgling.draw_tree(rctree)
out.movement_arrow((1,1,1,0), (1,1,0))
out.box_constituent((1,1,0))
rctree2 = ("DP", ("D", "a"), ("NP", ("N", "doctor"),
                             ("CP", ("DP", "who(i)"),
                                    ("TP", ("DP", "cats"),
                                             ("VP", ("V", "like"), ("DP", "t(i)"))))))
out2 = svgling.draw_tree(rctree2)
out2.movement_arrow((1,1,1,1,1), (1,1,0))
out2.box_constituent((1,1,0))
svgling.figure.SideBySide(out, out2)
```

## Relative clause examples

```python
%%lamb
||Alfonso|| = Alfonso_e
||is_|| = L f_<e,t> : f
||a|| = L f_<e,t> : f
||doctor|| = L x_e : Doctor(x)
||likes|| = L y_e : L x_e : Likes(x,y)
||cats|| = CATS_e # (not a very complete analysis!)
```

```python
from lamb.lang import IndexedPronoun, Trace, Binder
himself = lamb.lang.IndexedPronoun.index_factory("himself", tp("e"))
himself(1) # we set aside the presupposition for example purposes
```

A subject relative clause example:

```python
rc1 = doctor * (Binder(10, name="who") * (Trace(10) * (likes * cats)))
rc1
```

```python
display(rc1.source_tree())
rc1.tree()
```

```python
rc2 = doctor * (Binder(10, name="who") * (cats * (likes * Trace(10))))
rc2
```

```python
display(rc2.source_tree())
rc2.tree()
```

```python
rc3 = doctor * (Binder(10, name="who") * (Trace(10) * (likes * himself(10))))
rc3
```

```python
display(rc3.source_tree())
rc3.tree()
```

```python
him = lamb.lang.IndexedPronoun.index_factory("him", tp("e"))
rc3 = doctor * (Binder(10, name="who") * (Trace(10) * (likes * himself(11))))
rc3
```
