```python
# #light-source-example
from fdray import *

t = Texture(
    Pigment("brick", Color("white"), Color((0.8, 0.6, 0.4))).scale(0.1),
    Normal("brick", Normal("bumps", 0.2), Normal("granite", 0.3)).scale(0.1),
)
Scene(
    Camera(0, 10, look_at=(0, 0, 0.5), view_scale=1.5, distance=3),
    LightSource((0, 0, 4), "white", from_camera=False),
    Sphere("z", 0.4, Color((0.5, 0.7, 0.7))),
    Box(-1, 1).translate(0, 0, -1).add(t),
).render(200, 200)
```
