# Constructive Solid Geometry

```python
from fdray import *


def render(*args):
    return Scene(
        Camera(30, 30, view_scale=2),
        LightSource((1, 30, 40), "white"),
        Cylinder((0, 0, 0), (3, 0, 0), 0.02, Color("red")),
        Cylinder((0, 0, 0), (0, 3, 0), 0.02, Color("green")),
        Cylinder((0, 0, 0), (0, 0, 3), 0.02, Color("blue")),
        *args,
    ).render(width=300, height=300)
```

## Union

```python
s = Sphere(0, 0.8, Color("yellow", 0.5)).finish(phong=0.5).translate("0.5*y")
c = Cube(0, 2, Color("cyan", 0.5)).translate("-0.5*y")
render(s + c)
```

## Difference

```python
render(s - c)
```

```python
render(c - s)
```

## Intersection

```python
render(s * c)
```

## Merge

```python
render(s | c)
```
