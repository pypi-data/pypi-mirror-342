```python
endpoint = lambda slug: f'http://proj-finance-backend.onrender.com/financial-ratios/{slug}'
```

```python
def get(slug='microsoft'):
    res     = pd.read_json(endpoint(slug))
    Raw     = pd.DataFrame([dict(x) for x in res['Raw']])
    Ratios  = pd.DataFrame([dict(x) for x in res['Ratios']])
    return Raw, Ratios

Raw, Ratios = get()
Ratios
```
